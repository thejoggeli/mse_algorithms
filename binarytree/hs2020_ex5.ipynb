```python
values = [
    15, 
    6, 18,
    3, 7, 17, 20,
    2, 4, None, 13, *[None]*4,
    *[None]*4, 9,
]

tree = BinaryTree()
tree.build(values)

print("depth:", tree.depth)
print("minimum:", tree.minimum())
print("find:", tree.find(55))

tree.draw()

print("delete 18")
tree.delete(tree.find(18))
tree.draw()

print("insert 1")
tree.insert(1)
tree.draw()

print("insert 0")
tree.insert(0)
tree.draw()

print("delete maximum")
tree.delete(tree.maximum())
tree.draw()
```
