```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from MyGraph import Graph, Node, Edge
```

```python
# create graph "manually"
graph = Graph()
graph.verbose = False

# add nodes
chars = "abcdef"
for i in range(len(chars)):
    graph.add_node(chars[i])

# add edges
graph.add_edge("a", "b", 1.0)
graph.add_edge("a", "c", 5.0)
graph.add_edge("b", "d", 3.0)
graph.add_edge("b", "e", 1.0)
graph.add_edge("c", "f", 1.0)
graph.add_edge("d", "e", 3.0)
graph.add_edge("e", "c", 1.0)
graph.add_edge("e", "f", 2.0)

# build weight matrix
graph.build_matrix()

# get a node
node = graph.get_node("e")
print(node)
print("relatives:", len(node.relatives))
print("children:", len(node.children))
print("parents:", len(node.parents))

# get an edge
edge = graph.get_edge("a", "c")
print(edge)

# print and drawgraph
# print(graph.matrix)
graph.draw()
    
```

```python
# dijkstra
path = graph.dijkstra(source="a", target="f", verbose=True)
path_names = [node.name for node in path]
print("dijkstra path: " + str(path_names))

graph.draw(path=path)
```
