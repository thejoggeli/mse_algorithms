```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
def f(n):
    t = 0
    for i in range(1, n, 2):
        for j in range(0, i, 1):
            for k in range(0, 3, 1):
                t += 1
    return t
```

```python
x = np.arange(0, 500, 5)
y = np.array([f(n) for n in x])

plt.rcParams['figure.figsize'] = [6, 3]
fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid(True)
plt.show()
```

```python
def g(x, a, b, c, d):
    return d*x**3 + c*x**2 + b*x + a

opt, cov = curve_fit(g, x, y)
(a, b, c, d) = opt

print(f"a (x^0): {a:.3f}")
print(f"b (x^1): {b:.3f}")
print(f"c (x^2): {c:.3f}")
print(f"d (x^3): {d:.3f}")

y_fit = np.array([g(n, *opt) for n in x])
y_asymptote = c*x*x

plt.rcParams['figure.figsize'] = [10, 6]
fig, ax = plt.subplots()
ax.plot(x, y, lw=6, label="f(n)")
ax.plot(x, y_fit, lw=3, label="fit")
ax.plot(x, y_asymptote, lw=1, ls="--", label="asymptote", color="black")
ax.grid(True)
ax.legend()
plt.show()
```
